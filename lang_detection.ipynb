Как человек, вы можете легко определить, на каком из известных вам языков написан текст. Но компьютеру это дается не так легко. Например, Google Translate включает модель машинного обучения для определения языка, на случай, если пользователь точно не знает, какой язык хочет перевести.

image.png

Набор данных, который мы используем, содержит данные о 22 популярных языках и содержит 1000 предложений на каждом из языков, поэтому он будет подходящим набором данных для обучения модели определения языка с помощью машинного обучения. **Итак, давайте обучим модель обнаружения языка с помощью машинного обучения с использованием Python.**

````python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/dataset.csv")
data.head()
````

Разделим набор данных на факторы и результативный признак.

````python
x = np.array(data["Text"])
y = np.array(data["language"])

cv = CountVectorizer()
X = cv.fit_transform(x)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.33, 
                                                    random_state=42)
````

Поскольку это проблема многоклассовой классификации, мы будем использовать алгоритм мультиномиального наивного Байеса.

````python
model = MultinomialNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
````

Теперь давайте воспользуемся нашей моделью:

````python
text = "देखकर अच्छा लगता है"
data = cv.transform([text]).toarray()
output = model.predict(data)
print(output)
````

**Итак, наша модель работает отлично!**
